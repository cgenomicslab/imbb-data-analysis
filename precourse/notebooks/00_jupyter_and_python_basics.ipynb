- **Markdown cells** — contain formatted text (like this one).

### Essential keyboard shortcuts

Jupyter has two modes:

- **Edit mode** (green border): you are typing inside a cell. Press `Escape` to leave.
- **Command mode** (blue border): you can navigate between cells. Press `Enter` to edit the selected cell.

**Shortcuts in command mode** (press `Escape` first):

| Shortcut | Action |
|----------|--------|
| `Shift+Enter` | Run cell and move to next |
| `Ctrl+Enter` | Run cell and stay |
| `A` | Insert cell above |
| `B` | Insert cell below |
| `DD` | Delete cell (press D twice) |
| `M` | Convert cell to Markdown |
| `Y` | Convert cell to Code |
| `Z` | Undo cell operation |

**Shortcuts in edit mode:**

| Shortcut | Action |
|----------|--------|
| `Shift+Enter` | Run cell |
| `Tab` | Autocomplete |
| `Ctrl+/` | Comment/uncomment line |

**Try it:** press `Escape`, then `B` to insert a new cell below this one. Type `print("it works")` and run it with `Shift+Enter`.

## 2. Variables

A variable is a name that refers to a value. You create it with `=`.

```python
gene_name = "BRCA1"
expression_level = 245.7
sample_count = 50
is_significant = True

print("Gene:", gene_name)
print("Expression:", expression_level)
print("Samples:", sample_count)
print("Significant?", is_significant)
```

Variable names can contain letters, numbers, and underscores. They cannot start with a number. Names are case-sensitive: `Gene` and `gene` are different.

## 3. Data Types

Every value in Python has a **type**:

- `str` — text (string), written in quotes: `"BRCA1"`
- `int` — whole number: `50`
- `float` — decimal number: `245.7`
- `bool` — logical: `True` or `False`

```python
print(type("BRCA1"))
print(type(50))
print(type(245.7))
print(type(True))
```

You can convert between types when it makes sense.

```python
print(int("42"))     # string to integer
print(float(10))     # integer to float
print(str(3.14))     # number to string
print(int(9.8))      # float to integer (truncates, does not round)
```

## 4. Arithmetic

```python
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Integer division:", a // b)
print("Remainder:", a % b)
print("Power:", a ** b)
```

Division (`/`) always returns a float. Integer division (`//`) returns a whole number.

Use `round()` to control decimal places.

```python
result = 10 / 3
print("Raw:", result)
print("Rounded:", round(result, 2))
```

### Comparisons

Comparisons return `True` or `False`.

```python
x = 10

print("x == 10:", x == 10)   # equal to
print("x != 5:",  x != 5)    # not equal to
print("x > 3:",   x > 3)     # greater than
print("x < 3:",   x < 3)     # less than
print("x >= 10:", x >= 10)   # greater than or equal to
```

`=` assigns a value. `==` checks equality. They are different.

## 5. Strings

Strings are sequences of characters. You can use single or double quotes.

```python
gene = "BRCA1"
organism = 'Homo sapiens'

# Length
print("Length:", len(gene))

# Concatenation (joining strings)
label = gene + " (" + organism + ")"
print(label)

# Case conversion
print("Upper:", "atcg".upper())
print("Lower:", "ATCG".lower())
```

### Indexing strings

Individual characters can be accessed by position. **Python counts from 0.**

```python
seq = "ATCGATCG"

print("First:", seq[0])
print("Second:", seq[1])
print("Last:", seq[-1])
print("First three:", seq[0:3])
```

### Useful string methods

```python
# Split into parts
header = "gene_id;gene_name;chromosome"
parts = header.split(";")
print(parts)

# Join parts back
print(", ".join(parts))

# Replace characters
print("ATCG".replace("T", "U"))

# Strip whitespace from edges
print("  messy data  ".strip())

# Check if a substring exists
print("ATG" in "ATGATCG")

# Count occurrences
print("ATGATGATG".count("ATG"))
```
