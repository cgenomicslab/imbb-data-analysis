## 1. Lists

A list is an ordered collection of items. You create one with square brackets.

```python
genes = ["BRCA1", "TP53", "EGFR", "MYC", "KRAS"]
expression_values = [245.7, 128.3, 456.1, 89.4, 312.8]

print("Genes:", genes)
print("Number of genes:", len(genes))
```

### Indexing

Access individual elements by position. **Python counts from 0.**

```python
print("First gene:", genes[0])
print("Second gene:", genes[1])
print("Last gene:", genes[-1])
```

### Slicing

Extract a sub-list with `[start:stop]`. The start is included, the stop is excluded.

```python
print("First three:", genes[0:3])
print("From index 2 onward:", genes[2:])
print("Up to index 2:", genes[:2])
```

### Modifying lists

```python
# Append: add to the end
genes.append("PIK3CA")
print("After append:", genes)

# Change an element
genes[0] = "BRCA2"
print("After change:", genes)
```

### Useful list operations

```python
numbers = [5, 2, 8, 1, 9, 3]

print("Length:", len(numbers))
print("Sum:", sum(numbers))
print("Min:", min(numbers))
print("Max:", max(numbers))
print("Sorted:", sorted(numbers))
print("Is 8 in the list?", 8 in numbers)
```

## 2. Dictionaries

A dictionary stores **key-value pairs**. You look up a value by its key, like looking up a word in a real dictionary. Dictionaries use curly braces `{}`.

```python
gene_info = {
    "name": "BRCA1",
    "chromosome": 17,
    "length": 81189,
    "protein_coding": True
}

print(gene_info)
```

```python
# Access by key
print("Gene name:", gene_info["name"])
print("Chromosome:", gene_info["chromosome"])
```

```python
# Add a new key
gene_info["function"] = "DNA repair"

# Modify an existing key
gene_info["length"] = 81188

print(gene_info)
```

### Mapping names to values

A common pattern: use a dictionary to associate names with numbers.

```python
expression = {
    "BRCA1": 245.7,
    "TP53": 128.3,
    "EGFR": 456.1,
    "MYC": 89.4
}

print("TP53 expression:", expression["TP53"])
print("All keys:", list(expression.keys()))
print("All values:", list(expression.values()))
```

## 3. Conditionals: if / elif / else

Conditionals let your code make decisions. The indented block below the `if` only runs when the condition is `True`.

**Important:** Python uses **indentation** (4 spaces) to define blocks of code. This is not optional.

```python
p_value = 0.03

if p_value < 0.05:
    print("Significant")
else:
    print("Not significant")
```

```python
fold_change = 3.5

if fold_change > 2:
    print("Strongly upregulated")
elif fold_change > 1:
    print("Mildly upregulated")
else:
    print("Downregulated or no change")
```

Use `and`, `or`, and `not` to combine conditions.

```python
p_value = 0.01
fold_change = 2.5

if p_value < 0.05 and fold_change > 2:
    print("Significant and strongly upregulated")
```

## 4. For Loops

A `for` loop repeats a block of code once for each item in a sequence.

```python
genes = ["BRCA1", "TP53", "EGFR"]

for gene in genes:
    print("Processing:", gene)
```

### Loops with numbers: `range()`

`range(n)` produces numbers from 0 to n-1.

```python
for i in range(5):
    print("i =", i)
```

### Combining loops and conditionals

```python
expression_values = [245.7, 128.3, 456.1, 89.4, 312.8]
threshold = 200

for value in expression_values:
    if value > threshold:
        print(value, "is above threshold")
```

### Building a new list with a loop

A common pattern: start with an empty list and `append()` to it inside a loop.

```python
expression_values = [245.7, 128.3, 456.1, 89.4, 312.8]
high_expression = []

for value in expression_values:
    if value > 200:
        high_expression.append(value)

print("High expression values:", high_expression)
```

### Looping through a dictionary

```python
expression = {"BRCA1": 245.7, "TP53": 128.3, "EGFR": 456.1}

for gene, value in expression.items():
    print(gene, "->", value)
```

### `enumerate()`: get both index and value

```python
genes = ["BRCA1", "TP53", "EGFR"]

for i, gene in enumerate(genes):
    print("Index", i, ":", gene)
```
