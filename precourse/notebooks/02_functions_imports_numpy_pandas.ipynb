## 1. Functions

A function is a named block of code that you can run whenever you need it. You define a function with `def` and call it by name.

````python
def greet(name):
    print("Hello,", name)

greet("Alice")
greet("Bob")
````

### Return values

Most useful functions **return** a result that you can store in a variable.

````python
def fold_change(treated, control):
    return treated / control

fc = fold_change(360, 120)
print("Fold change:", fc)
````

### Default parameter values

Parameters can have a default. If the caller does not provide that argument, the default is used.

````python
def is_significant(p_value, threshold=0.05):
    return p_value < threshold

print(is_significant(0.03))              # uses default 0.05
print(is_significant(0.08))              # uses default 0.05
print(is_significant(0.08, threshold=0.1))  # overrides default
````

### Functions can return multiple values

````python
def describe_list(values):
    return min(values), max(values), sum(values) / len(values)

data = [245.7, 128.3, 456.1, 89.4, 312.8]
minimum, maximum, mean = describe_list(data)

print("Min:", minimum)
print("Max:", maximum)
print("Mean:", round(mean, 1))
````

### A practical example

````python
def gc_content(sequence):
    """Calculate GC content of a DNA sequence."""
    sequence = sequence.upper()
    gc_count = sequence.count("G") + sequence.count("C")
    return gc_count / len(sequence)

print("GC content:", round(gc_content("ATCGATCGATCG"), 3))
print("GC content:", round(gc_content("GCGCGCATATAT"), 3))
````

The text in triple quotes (`"""..."""`), right after `def`, is called a **docstring**. It describes what the function does.

## 2. Imports

Python comes with many pre-built libraries (also called **modules**). To use them, you must **import** them first.

````python
import math

print("Square root of 16:", math.sqrt(16))
print("Log base 2 of 8:", math.log2(8))
print("Pi:", math.pi)
````

### Different ways to import

````python
# Import specific functions
from math import sqrt, log2
print(sqrt(25))
print(log2(16))
````

````python
# Import with an alias (a short nickname)
import math as m
print(m.sqrt(36))
````

Aliases are a convention. The standard ones in scientific Python:

## 3. NumPy: Numerical Arrays

NumPy is the library for numerical computing. Its core object is the **array** — like a list, but designed for math. The key advantage: operations apply to **every element at once**, without writing a loop.

````python
import numpy as np

expression = np.array([245.7, 128.3, 456.1, 89.4, 312.8])
print("Array:", expression)
````

````python
# Element-wise operations — no loop needed
print("Doubled:", expression * 2)
print("Plus 10:", expression + 10)
print("Log2:", np.log2(expression))
````

````python
# Summary statistics
print("Mean:", round(np.mean(expression), 2))
print("Std dev:", round(np.std(expression), 2))
print("Min:", np.min(expression))
print("Max:", np.max(expression))
````

````python
# Boolean indexing: filter elements by a condition
print("Above 200?", expression > 200)
print("Values above 200:", expression[expression > 200])
````

Pandas (below) is built on top of NumPy. When you work with a column of numbers in a pandas DataFrame, it is stored as a NumPy array internally.

## 4. Pandas: DataFrames

Pandas is the central library for working with tabular data in Python. If you have used spreadsheets, a pandas **DataFrame** is the Python equivalent: data organized in rows and columns.

### Filtering rows

You can filter a DataFrame by applying a condition to a column.

````python
# Beaches with temperature above 24
warm = beaches[beaches["water_temp"] > 24]
print("Warm beaches:", len(warm))
warm.head()
````

````python
# Beaches in a specific region
chania = beaches[beaches["region"] == "Chania"]
print("Beaches in Chania:", len(chania))
````

Combine conditions with `&` (and) or `|` (or). Each condition must be in parentheses.

````python
result = beaches[(beaches["water_temp"] > 24) & (beaches["region"] == "Chania")]
print("Warm Chania beaches:", len(result))
````
