## 1. Printing and Variables

In R, you can print a value by simply typing it, or by using `print()`. Variables are created with `=`.

````r
gene_name = "BRCA1"
expression = 245.7
sample_count = 50
is_significant = TRUE

print(gene_name)
print(expression)
print(sample_count)
print(is_significant)
````

In Jupyter, the last expression in a cell is displayed automatically, just like in Python.

````r
6 * 7
````

To print a label and a value together, use `paste()`.

````r
print(paste("Gene:", gene_name))
print(paste("Expression:", expression))
````

### Data types

R has similar basic types to Python, with slightly different names:

| Python | R | Example |
|--------|---|--------|
| `str` | `character` | `"BRCA1"` |
| `int` / `float` | `numeric` | `245.7` |
| `bool` | `logical` | `TRUE` / `FALSE` |

Use `class()` to check the type of a value (equivalent to Python's `type()`).

````r
class("BRCA1")
class(245.7)
class(TRUE)
````

Note: R uses `TRUE` and `FALSE` (all caps), not `True` and `False`.

## 2. Arithmetic and Comparisons

Arithmetic works the same as in Python.

````r
a = 10
b = 3

print(paste("Addition:", a + b))
print(paste("Subtraction:", a - b))
print(paste("Multiplication:", a * b))
print(paste("Division:", a / b))
print(paste("Integer division:", a %/% b))
print(paste("Remainder:", a %% b))
print(paste("Power:", a ^ b))
````

Minor differences from Python: power is `^` instead of `**`, integer division is `%/%`, remainder is `%%`.

Comparisons work the same way.

````r
x = 10

x == 10
x != 5
x > 3
x < 3
````

Logical operators: `&` (and), `|` (or), `!` (not) — same symbols as Python uses inside pandas.

## 3. Vectors

The most fundamental data structure in R is the **vector** — an ordered collection of values of the same type. You create one with `c()` (short for "combine").

Vectors are roughly equivalent to Python lists or NumPy arrays.

````r
genes = c("BRCA1", "TP53", "EGFR", "MYC", "KRAS")
expression = c(245.7, 128.3, 456.1, 89.4, 312.8)

print(genes)
print(expression)
print(paste("Length:", length(expression)))
````

### Indexing

**Important:** R counts from 1, not from 0.

````r
print(paste("First gene:", genes[1]))
print(paste("Third gene:", genes[3]))

# Slicing: select elements 2 through 4
print(genes[2:4])
````

### Element-wise operations

Like NumPy arrays, R vectors support element-wise operations without writing a loop.

````r
expression = c(245.7, 128.3, 456.1, 89.4, 312.8)

# Multiply every element by 2
expression * 2
````

````r
# Log2 of every element
log2(expression)
````

### Summary statistics

````r
print(paste("Mean:", mean(expression)))
print(paste("Median:", median(expression)))
print(paste("Std dev:", round(sd(expression), 2)))
print(paste("Min:", min(expression)))
print(paste("Max:", max(expression)))
````

### Boolean indexing

Filter a vector using a condition — same idea as NumPy and pandas.

````r
# Which values are above 200?
expression > 200
````

````r
# Select only those values
expression[expression > 200]
````

## 4. Data Frames

An R data frame is a table of data — equivalent to a pandas DataFrame. Each column is a vector.

````r
df = data.frame(
    gene = c("BRCA1", "TP53", "EGFR", "MYC", "KRAS"),
    chromosome = c(17, 17, 7, 8, 12),
    expression = c(245.7, 128.3, 456.1, 89.4, 312.8)
)

df
````

### Exploring a data frame

````r
# First few rows
head(df)
````

````r
# Dimensions: rows, columns
dim(df)
````

````r
# Column names
colnames(df)
````

````r
# Structure (types of each column)
str(df)
````

````r
# Summary statistics
summary(df)
````

### Selecting columns

Use `$` to access a column by name.

````r
df$gene
````

````r
df$expression
````

````r
mean(df$expression)
````

### Filtering rows

````r
# Rows where expression is above 200
df[df$expression > 200, ]
````

````r
# Rows where chromosome is 17
df[df$chromosome == 17, ]
````

Note the syntax: `df[rows, columns]`. The comma is required — leaving the column part empty means "all columns".

## 5. Functions

Functions in R are defined with `function()` and assigned to a name.

````r
fold_change = function(treated, control) {
    return(treated / control)
}

fold_change(360, 120)
````

````r
gc_content = function(sequence) {
    sequence = toupper(sequence)
    chars = strsplit(sequence, "")[[1]]
    gc = sum(chars == "G") + sum(chars == "C")
    return(gc / nchar(sequence))
}

gc_content("ATCGATCG")
````

## Quick Reference: Python vs R

| Concept | Python | R |
|---------|--------|---|
| Assignment | `x = 5` | `x = 5` |
| Print | `print("hello")` | `print("hello")` |
| Type check | `type(x)` | `class(x)` |
| Boolean values | `True` / `False` | `TRUE` / `FALSE` |
| List / Vector | `[1, 2, 3]` | `c(1, 2, 3)` |
| Length | `len(x)` | `length(x)` |
| Indexing | `x[0]` (starts at 0) | `x[1]` (starts at 1) |
| Element-wise ops | `np.log2(x)` | `log2(x)` |
| Create data frame | `pd.DataFrame({...})` | `data.frame(...)` |
| Load CSV | `pd.read_csv("f.csv")` | `read.csv("f.csv")` |
| First rows | `df.head()` | `head(df)` |
| Dimensions | `df.shape` | `dim(df)` |
| Select column | `df["col"]` | `df$col` |
| Filter rows | `df[df["col"] > 5]` | `df[df$col > 5, ]` |
| Define function | `def f(x): return x` | `f = function(x) { return(x) }` |
| Import / Load | `import pandas as pd` | `library(ggplot2)` |
| Install package | `pip install x` | `install.packages("x")` |
| Power | `x ** 2` | `x ^ 2` |
