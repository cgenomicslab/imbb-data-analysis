## 1. Data Visualization with Seaborn

**Seaborn** makes beautiful statistical plots easy. We'll skip matplotlib complexity and focus on seaborn.

### Setup: Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Set style for prettier plots
sns.set_style('whitegrid')

print('✅ Libraries loaded!')
```

### Scatter Plot: Relationship Between Two Variables

```python
# Create sample data
gene_length = [1500, 2300, 800, 3200, 1200, 2800, 1800]
expression = [120, 180, 60, 250, 90, 210, 140]

# Create scatter plot
plt.figure(figsize=(8, 5))
sns.scatterplot(x=gene_length, y=expression, s=100)
plt.xlabel('Gene Length (bp)')
plt.ylabel('Expression Level')
plt.title('Gene Length vs Expression')
plt.show()
```

### Histogram: Distribution of Values

```python
# Expression values for many genes
expression_values = np.random.normal(150, 40, 100)

plt.figure(figsize=(8, 5))
sns.histplot(expression_values, bins=20, kde=True)
plt.xlabel('Expression Level')
plt.ylabel('Frequency')
plt.title('Distribution of Gene Expression')
plt.show()
```

### Box Plot: Compare Groups

```python
# Simulated expression in 3 conditions
data = {
    'condition': ['Normal']*20 + ['Treatment_A']*20 + ['Treatment_B']*20,
    'expression': list(np.random.normal(100, 15, 20)) + 
                  list(np.random.normal(150, 20, 20)) + 
                  list(np.random.normal(80, 12, 20))
}
df_temp = pd.DataFrame(data)

plt.figure(figsize=(8, 5))
sns.boxplot(data=df_temp, x='condition', y='expression')
plt.ylabel('Expression Level')
plt.title('Expression Across Conditions')
plt.show()
```

## 2. Introduction to Pandas DataFrames

**Pandas** is the most important library for data analysis. DataFrames are like Excel spreadsheets in Python.

### Loading Our First Dataset: Cretan Beaches!

```python
# Load the dataset
beaches = pd.read_csv('../data/cretan_beaches.csv')

# Display first few rows
print('First 5 beaches:')
beaches.head()
```

```python
# Basic information about the dataset
print(f'Number of beaches: {len(beaches)}')
print(f'Number of columns: {len(beaches.columns)}')
print(f'\nColumn names:\n{beaches.columns.tolist()}')
```

```python
# Summary statistics
beaches.describe()
```

### Selecting Columns

```python
# Select a single column
print('Beach names:')
print(beaches['beach_name'])

# Select multiple columns
print('\nBeach info:')
beaches[['beach_name', 'region', 'beach_length_m']].head()
```

### Selecting Rows: .loc and .iloc

```python
# Select by index number (.iloc)
print('First beach:')
print(beaches.iloc[0])

# Select multiple rows
print('\nFirst 3 beaches:')
beaches.iloc[0:3]
```

### Basic Operations on DataFrames

```python
# Calculate mean
mean_tourists = beaches['avg_tourists_per_day'].mean()
print(f'Average tourists per day: {mean_tourists:.0f}')

# Find maximum
max_temp = beaches['water_temp_celsius'].max()
print(f'Highest water temperature: {max_temp}°C')

# Count unique values
print(f'\nRegions in dataset:')
print(beaches['region'].value_counts())
```

## 3. Filtering Data (Boolean Indexing)

**This is very important!** Boolean indexing lets you filter rows based on conditions.

### Simple Filtering

```python
# Find beaches in Chania
chania_beaches = beaches[beaches['region'] == 'Chania']
print(f'Beaches in Chania: {len(chania_beaches)}')
print(chania_beaches[['beach_name', 'beach_type']].head())
```

```python
# Find beaches with high tourism
crowded = beaches[beaches['avg_tourists_per_day'] > 300]
print(f'\nCrowded beaches (>300 tourists/day): {len(crowded)}')
crowded[['beach_name', 'avg_tourists_per_day']]
```

### Multiple Conditions: AND (&) and OR (|)

```python
# Beaches in Chania AND with low tourism
quiet_chania = beaches[(beaches['region'] == 'Chania') & 
                       (beaches['avg_tourists_per_day'] < 100)]
print('Quiet beaches in Chania:')
print(quiet_chania[['beach_name', 'avg_tourists_per_day']])
```

```python
# Beaches with Kri-kri OR monk seals
wildlife = beaches[(beaches['kri_kri_sightings'] > 0) | 
                   (beaches['monk_seal_sightings'] > 0)]
print(f'\nBeaches with wildlife: {len(wildlife)}')
wildlife[['beach_name', 'kri_kri_sightings', 'monk_seal_sightings']].head()
```

## 4. Grouping and Aggregation

**groupby()** lets you split data into groups and calculate statistics for each group.

```python
# Average tourists by region
by_region = beaches.groupby('region')['avg_tourists_per_day'].mean()
print('Average tourists per day by region:')
print(by_region)
```

```python
# Multiple statistics at once
region_stats = beaches.groupby('region')['water_temp_celsius'].agg(['mean', 'min', 'max'])
print('\nWater temperature by region:')
print(region_stats)
```

```python
# Count by beach type
beach_types = beaches.groupby('beach_type').size()
print('\nNumber of beaches by type:')
print(beach_types)
```

## 5. Plotting Pandas Data

Combining pandas with seaborn makes powerful visualizations easy!

```python
# Scatter plot: tourists vs water temperature
plt.figure(figsize=(8, 5))
sns.scatterplot(data=beaches, x='water_temp_celsius', y='avg_tourists_per_day', 
                hue='region', s=100)
plt.xlabel('Water Temperature (°C)')
plt.ylabel('Average Tourists per Day')
plt.title('Beach Popularity vs Water Temperature')
plt.show()
```

```python
# Box plot by region
plt.figure(figsize=(10, 5))
sns.boxplot(data=beaches, x='region', y='water_temp_celsius')
plt.ylabel('Water Temperature (°C)')
plt.title('Water Temperature by Region')
plt.show()
```

```python
# Bar plot: average by beach type
type_tourists = beaches.groupby('beach_type')['avg_tourists_per_day'].mean()

plt.figure(figsize=(8, 5))
type_tourists.plot(kind='bar')
plt.ylabel('Average Tourists per Day')
plt.title('Tourism by Beach Type')
plt.xticks(rotation=45)
plt.show()
```
