## 2. Lists: Storing Multiple Values

Lists are ordered collections that can store multiple items. Think of a list as a container with numbered slots.

```python
# Creating a list
gene_names = ["BRCA1", "TP53", "EGFR", "MYC", "KRAS"]
expression_values = [245.7, 128.3, 456.1, 89.4, 312.8]

print("Genes:", gene_names)
print("Expression:", expression_values)
```

### Accessing List Elements (Indexing)

**Important:** Python counts from 0, not 1!

```python
# Indexing starts at 0
print("First gene:", gene_names[0])   # BRCA1
print("Second gene:", gene_names[1])  # TP53
print("Last gene:", gene_names[-1])   # KRAS (negative index counts from end)

# Length of a list
print("Number of genes:", len(gene_names))
```

### List Slicing: Getting Subsets

Slicing lets you extract parts of a list: `list[start:end]`
- Includes `start` index
- Excludes `end` index

```python
# Slicing examples
print("First 3 genes:", gene_names[0:3])   # indices 0, 1, 2
print("From index 2 onward:", gene_names[2:])  # index 2 to end
print("Up to index 3:", gene_names[:3])    # start to index 2
print("Last 2 genes:", gene_names[-2:])    # last 2 elements
```

### Adding Elements to Lists: The `append()` Method

**This is important!** `append()` adds a new element to the end of a list.

```python
# Start with an empty list
samples = []
print("Initial list:", samples)

# Add elements one by one
samples.append("Sample_1")
samples.append("Sample_2")
samples.append("Sample_3")

print("After appending:", samples)
```

## 3. Dictionaries: Key-Value Pairs

Dictionaries store data as key-value pairs. They're like a phonebook: you look up a name (key) to get a phone number (value).

```python
# Creating a dictionary
gene_info = {
    "name": "BRCA1",
    "chromosome": 17,
    "length": 81189,
    "protein_coding": True
}

print(gene_info)
```

### Accessing Dictionary Values

### Adding/Modifying Dictionary Entries

```python
# Add a new key-value pair
gene_info["function"] = "DNA repair"

# Modify existing value
gene_info["length"] = 81188  # corrected length

print(gene_info)
```

## 4. Conditionals: Making Decisions

Conditionals let your code make decisions based on conditions.

```python
# Simple if statement
p_value = 0.03
threshold = 0.05

if p_value < threshold:
    print("Result is statistically significant!")
else:
    print("Result is not significant.")
```

### Multiple Conditions: if-elif-else

```python
# Classify expression level
expression = 180.5

if expression > 200:
    print("High expression")
elif expression > 100:
    print("Medium expression")
else:
    print("Low expression")
```

### Comparison Operators

- `==` equal to
- `!=` not equal to
- `<` less than
- `>` greater than
- `<=` less than or equal
- `>=` greater than or equal

```python
# Examples
fold_change = 2.5

print("Fold change > 2:", fold_change > 2)
print("Fold change == 2.5:", fold_change == 2.5)
print("Fold change >= 3:", fold_change >= 3)
```

## 5. Loops: Repeating Actions

Loops let you repeat code multiple times without writing it over and over.

### For Loops: Iterate Over a List

```python
# Loop through a list
genes = ["BRCA1", "TP53", "EGFR"]

for gene in genes:
    print("Processing gene:", gene)
```

### Loops with Conditionals

```python
# Filter expression values
expression_values = [245.7, 128.3, 456.1, 89.4, 312.8]
threshold = 200

print("Highly expressed genes:")
for value in expression_values:
    if value > threshold:
        print("  Expression:", value)
```

### Building Lists with Loops

```python
# Create a new list based on conditions
expression_values = [245.7, 128.3, 456.1, 89.4, 312.8]
high_expression = []

for value in expression_values:
    if value > 200:
        high_expression.append(value)  # Here's append!

print("High expression values:", high_expression)
```

## 6. Functions: Reusable Code

Functions are named blocks of code that you can use multiple times. They help organize your code and avoid repetition.

### Defining a Simple Function

```python
# Define a function
def calculate_mean(first, second):
    """Calculate the mean of two numbers"""
    mean_of_two = (first + second) / 2
    return mean_of_two

# Use the function
a = 10
b = 20
mean_value = calculate_mean(a, b)
print("The mean of", a, "and", b, "is", mean_value)
```

```python
# Define a function
def calculate_gc_content(sequence):
    """Calculate GC content of a DNA sequence."""
    gc_count = sequence.count('G') + sequence.count('C')
    gc_percentage = (gc_count / len(sequence)) * 100
    return gc_percentage

# Use the function
dna = "ATGCGCTAGCTAGCGCTA"
gc = calculate_gc_content(dna)
print("GC content:", round(gc, 1), "%")
```

### Function with Multiple Parameters

```python
def is_significant(p_value, threshold=0.05):
    """Check if a p-value is statistically significant."""
    return p_value < threshold

# Test the function
print("p=0.03 significant?", is_significant(0.03))
print("p=0.12 significant?", is_significant(0.12))
print("p=0.08 significant at 0.1?", is_significant(0.08, threshold=0.1))
```

### Why Functions Are Useful

```python
def normalize_expression(value, mean, std):
    """Z-score normalization."""
    return (value - mean) / std

# Normalize multiple values easily
mean_expr = 200
std_expr = 50

values = [180, 220, 150, 280]
normalized = []

for val in values:
    norm_val = normalize_expression(val, mean_expr, std_expr)
    normalized.append(norm_val)

print("Original:", values)
print("Normalized:", normalized)
```

## 8. Aside: Formatting Output

So far we have used `print("label:", value)` and `round()`. Python also has **f-strings** — a compact way to mix text and variables in a single string. Put `f` before the quote and wrap variables in `{}`:

```python
gene = "TP53"
expression = 128.3

# f-string: variables inside {}
print(f"Gene {gene} has expression {expression}")

# Control decimal places with :.Nf
ratio = 10 / 3
print(f"Ratio: {ratio:.2f}")
print(f"Ratio: {ratio:.4f}")

# Both styles work — use whichever is clearer
print("Ratio:", round(ratio, 2))  # same result, different syntax
```
